This notebook shows an example of how to use oa_data_module.py
With this file, user can compile EMG, AAC, fNIRS and Treadmill data into a single file, and perform the time synchronization

```python
from oa_data_module import OA_data
```

Create the module by specifying the trial number.

```python
abc = OA_data('302')
```

Fitting functions

this function will fit the module by applying EMG, ACC, fNIRs and Treadmill data. The defaut directory each dataframe is located is set as below
- 1. EMG/ACC: csv/
- 2. fNIRS: fNIRS_Walking data
- 3. TreadMill: Treadmill_data

To change the directory dataframe is located, simply pass them as a argument

```python
abc.fit_emg_acc()
abc.fit_fNIRS()
abc.fit_TMdata()
```

Synching functions

This function will perform the time synchronization.

```python
abc.time_synch()
```

synched_to_csv(trial_number)

This function will store time synched data into csv format
The argument is the trial number you want to save. Pass corresponding integer as an argument.
- 1: N1
- 2: N2
- 3: P1
- 4: P4

```python
#this will save following 4 files
#synched_302_emg_N1.csv, synched_302_acc_N1.csv, 
#synched302_fNIRS_N1.csv, synched302_treadmill_N1.csv
abc.synched_to_csv(1)
```

Compilation function

this function will compile dataframes we fitted before into a single file in HDF style

```python
#this will create file named "302.h5"
abc.save()
```
